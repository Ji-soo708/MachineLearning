```python
trans_li = [] #치환하고 넣을 리스트
finfintra_li=[] #마지막으로 치환한 문장들을 담는 리스트
sen_list1=[] #원본 문장들이 들어가는 리스트

#sen_list1=sen_list
for i in sen_list:
    temp = i.replace(".","") #온점 제거
    sen_list1.append(temp)
#print(sen_list1)
#sen_list1에 원본 문장들이 들어있습니다.
for i in sen_list1:
    i = i.split(" ") #띄어쓰기 제거
    trans_li=[]
    for k in i:
        str='' #변환한 문장 담는 변수, 초기화
        count=0 #카운트 변수
        for j in k: #문장에서 띄어쓰기 기준으로 단어를 나눠 단어별로 for문 돌리는 중 
            if(count==0): #첫음절만 B로 변환
                str+="B"
                count+=1
            else:
                str+="I" #나머지 음절은 I로 변환
        trans_li.append(str) #for문 다 돌아서 치환 문장(한개의 문장) 넣는 리스트에 저장
    #print(trans_li)
    #print(len(i))
    trans_split=''.join(trans_li) #마지막으로 for문 다 돌고 치환한 여러개의 단어들을 하나의 문장으로 합치기
    #print(trans_split)
    finfintra_li.append(trans_split) #찐마지막으로 치환한 문장'들'을 담는 리스트에 치환한 문장 하나씩 추가
print(finfintra_li) 
```
